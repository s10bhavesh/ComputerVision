```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

SIFT


```
def sift_transform(path):
  img = cv2.imread(path, 0)
  sift = cv2.xfeatures2d.SIFT_create()
  output = sift.detect(img,None)
  result = cv2.drawKeypoints(img, output, outImage = None, color=(255,0,0))
  return result

img = "/content/drive/My Drive/SEM2/Computer Vision/Lab/ship.png"
cv2_imshow(cv2.imread(img,0))
cv2_imshow(sift_transform(img))
```
