```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
#base_dir = '/content/drive/My Drive/SEM2/Computer Vision/Lab'
```

```python


img = cv2.imread('/content/drive/My Drive/SEM2/Computer Vision/Lab/opencvlogo.png',0)
#cv2_imshow('original image',img)
cv2_imshow(img)
img1 = cv2.medianBlur(img,5)
#cv2_imshow('blur image',img1)
cv2_imshow(img1)
```

```python
cimg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)
```

```python

circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,1,20,
                            param1=50,param2=30,minRadius=0,maxRadius=0)

circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)
    # draw the center of the circle
    cv2.circle(cimg,(i[0],i[1]),2,(0,0,255),3)

#cv2.imshow('detected circles',cimg)
cv2_imshow(cimg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
